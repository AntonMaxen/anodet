```python

from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
from matplotlib.figure import Figure

def getBatchVisImage(images, score_maps, thresh):
    
    fig_img, ax_img = plt.subplots(1, len(images), figsize=(12, 3*scores.shape[0]))
    canvas = FigureCanvas(fig_img)

    indent = 10
        
    
    for i in range(len(images)):
        ax_img[i].axis('off')
        
        frame = np.ones((score_maps[i].shape[0]+2*indent, score_maps[i].shape[0]+2*indent))

        mask = score_maps[i].copy()
        mask[mask > THRESH] = 255
        mask[mask <= THRESH] = 0

        resized = cv2.resize(images[i], (56,56), interpolation = cv2.INTER_AREA)
        vis_img = mark_boundaries(resized, mask, color=(1, 0, 0), mode='thick')
        ax_img[i].imshow(vis_img)

        ax_img[i].margins(0,0)
        
        
        
    canvas.draw()       # draw the canvas, cache the renderer


    image = np.fromstring(canvas.tostring_rgb(), dtype='uint8').reshape((int(fig_img.bbox.bounds[3]), int(fig_img.bbox.bounds[2]),3))
    image = image[470:600, 100:780]
    return image
        
```
